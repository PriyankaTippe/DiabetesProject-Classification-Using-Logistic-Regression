```python
# Importing Liabraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing dataset

ds = pd.read_csv("cleaned_diabetic_data.csv")
```

# Visualization by graph

```python
ds = ds.loc[:, ~ds.columns.str.contains('^Unnamed')]                # removing unnamed column
```

```python
ds.columns
```

```python
# Plotting histograms for each column in the dataset

ds.hist(bins=20, figsize=(15, 10), grid=False)
plt.suptitle('Histogram of Features', fontsize=16)
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = ds.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=16)
plt.show()
```

```python
# Create a boxplot for each feature

plt.figure(figsize=(15, 10))
sns.boxplot(data=ds, width=0.5, palette="Set2")
plt.title('Boxplot of Features', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot for the Outcome column

plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=ds, palette='Set1')
plt.title('Distribution of Outcome', fontsize=16)
plt.show()

# Pie Chart - Distribution of Outcome classes
plt.figure(figsize=(6, 6))
ds['Outcome'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title('Diabetes Outcome Distribution')
plt.ylabel('')
plt.show()
```

```python
# Violin plot for the Glucose column by Outcome
plt.figure(figsize=(8, 6))
sns.violinplot(x='Outcome', y='Glucose', data=ds, palette='muted')
plt.title('Violin Plot: Glucose by Outcome', fontsize=16)
plt.show()
```

```python
# Pairplot with regression lines
sns.pairplot(ds, hue='Outcome', kind='reg', height=2.5)
plt.suptitle('Pairplot with Regression Lines by Outcome', fontsize=16)
plt.show()
```
