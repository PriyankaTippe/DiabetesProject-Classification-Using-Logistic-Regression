```python
# Importing Liabraries

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing datasset

ds = pd.read_csv("diabetic.csv")
```

```python
for col in ds.columns:
     print(col)
     print(sorted(ds[col].unique()))
```

```python
ds.columns
```

```python
# Converts missing values using mean

columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in columns:
    mean = ds[col].mean()
    for i in ds.index:
        if ds.loc[i, col] == 0:
            ds.loc[i, col] = mean
```

```python
# Finding duplicated values if any

ds.duplicated().unique()
```

```python
ds.drop_duplicates(inplace=True)
```

```python
ds.dropna(inplace=True)
```

```python
ds.corr()
```

```python
ds.to_csv("cleaned_diabetic_data.csv")
```
